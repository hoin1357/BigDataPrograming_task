```python
import matplotlib.pyplot as plt

# 데이터 준비
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 4, 5]
y3 = [3, 2, 3, 4]
y4 = [4, 1, 2, 3]

# 그래프 그리기
plt.plot(x, y1, linestyle='-', label='Solid line')     # 실선
plt.plot(x, y2, linestyle='--', label='Dashed line')   # 점선
plt.plot(x, y3, linestyle=':', label='Dotted line')    # 점선(점)
plt.plot(x, y4, linestyle='-.', label='Dash-dot line') # 대시-점선

# 그래프 설정
plt.title("선 종류 (Line Styles in Matplotlib)")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# 범례 표시
plt.legend()

plt.show()
```

```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 4, 5]
y3 = [3, 2, 3, 4]
y4 = [4, 1, 2, 3]
y5 = [4, 5, 6, 8]

plt.plot(x, y1, marker='o', label='Circle')
plt.plot(x, y2, marker='^', label='Triangle_up')
plt.plot(x, y3, marker='s', label='Square')
plt.plot(x, y4, marker='*', label='Star')
plt.plot(x, y5, marker=',', label='Pixel')

plt.title("Marker Styles in Matplotlib")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.legend()
plt.show()
```

```python
import seaborn as sns

ax = sns.countplot(data=mtc, x='gear',
                   order=mtc.gear.value_counts().index)

for p in ax.patches:
    height = p.get_height()
    ax.text(x=(p.get_x() + p.get_width()/2.),
            y=height + .2,
            s=int(height),
            ha='center',
            size=12,
            color='blue')

ax.set_ylim(0, 17)
plt.show()
```

```python
iris = data('iris')
iris.Species.unique()
```

```python
n, bins, patches = plt.hist(iris['Sepal.Length'], bins=20, color='yellowgreen', edgecolor='white')

print(n)
print(bins)
print(patches)
```

```python
n, bins, patches = plt.hist(iris['Sepal.Length'], bins=20, color='yellowgreen', edgecolor='white')

for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))

plt.show()
```
